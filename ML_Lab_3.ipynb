Experiment 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

```python
mean_scores = df[numerical_cols].mean()
median_scores = df[numerical_cols].median()
mode_scores = df[numerical_cols].mode().iloc[0]

print("Mean:\n", mean_scores)
print("\nMedian:\n", median_scores)
print("\nMode:\n", mode_scores)
```

```python
print("Minimum:\n", df[numerical_cols].min())
print("\nMaximum:\n", df[numerical_cols].max())
print("\nSum:\n", df[numerical_cols].sum())
print("\nVariance:\n", df[numerical_cols].var())
print("\nStandard Deviation:\n", df[numerical_cols].std())
```

Experiment 2

```python
Q1 = df[numerical_cols].quantile(0.25)
Q2 = df[numerical_cols].quantile(0.50)
Q3 = df[numerical_cols].quantile(0.75)

print("First Quartile (Q1):\n", Q1)
print("\nSecond Quartile (Median / Q2):\n", Q2)
print("\nThird Quartile (Q3):\n", Q3)
```

Experiment 3

```python
correlation_matrix = df[numerical_cols].corr()
print("Correlation Matrix:\n", correlation_matrix)
```

```python
covariance_matrix = df[numerical_cols].cov()
print("Covariance Matrix:\n", covariance_matrix)
```

Experiment 4

```python
for col in numerical_cols:
    plt.hist(df[col], bins=10)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
for col in numerical_cols:
    plt.boxplot(df[col])
    plt.ylabel(col)
    plt.title(f"Boxplot of {col}")
    plt.show()
```

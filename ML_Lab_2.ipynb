Experiment 1 : Using Titanic Dataset

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
df.info()
```

```python
missing = df.isnull()

not_missing = df.notnull()

print("Missing values (True = Missing):")
print(missing.head())

print("\nNon-missing values (True = Not Missing):")
print(not_missing.head())

print("\nTotal missing values in each column:")
print(df.isnull().sum())
```

Experiment 2 : Remove rows with missing values

```python
print("Before cleaning:", df.shape)

df_dropna = df.dropna()

print("After cleaning:", df_dropna.shape)
```

Experiment 3 : Replace missing values

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)

df['Embarked'].fillna('Unknown', inplace=True)
df['Cabin'].fillna('Not Assigned', inplace=True)

df.isnull().sum()
```

Experiment 4 : Convert data types

```python
print("Before:")
print(df.dtypes)

df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')

print("\nAfter:")
print(df.dtypes)
```

Experiment 5 : Rename columns

```python
df.rename(columns={
    'PassengerId': 'Passenger_ID',
    'Pclass': 'Passenger_Class',
    'SibSp': 'Siblings_Spouses',
    'Parch': 'Parents_Children'
}, inplace=True)

df.head()
```

Experiment 6 : Fix inconsistent values

```python
df['Sex'] = df['Sex'].replace({
    'male': 'Male',
    'female': 'Female'
})

df['Sex'].unique()
```

Experiment 7 : Line plot & Bar plot

```python
import matplotlib.pyplot as plt

plt.plot(df['Passenger_ID'], df['Fare'])
plt.xlabel("Passenger ID")
plt.ylabel("Fare")
plt.title("Fare Trend Across Passengers")
plt.show()

df['Survived'].value_counts().plot(kind='bar')
plt.xlabel("Survived")
plt.ylabel("Count")
plt.title("Survival Count")
plt.show()
```

Experiment 8 : Scatter plot

```python
plt.scatter(df['Age'], df['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age vs Fare")
plt.show()
```

Experiment 9 : Histogram

```python
plt.hist(df['Age'], bins=10)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

Experiment 10 : Boxplot

```python
plt.boxplot(df['Fare'])
plt.ylabel("Fare")
plt.title("Fare Boxplot")
plt.show()
```
